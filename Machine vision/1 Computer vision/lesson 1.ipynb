```python
from imageai.Detection import ObjectDetection
from IPython.display import Image
import os
```

```python
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet() # Модель RetinaNet 
detector.setModelPath('resnet50_coco_best_v2.0.1.h5') 
detector.loadModel()
```

```python
Image(filename='Image 01.jpg')
```

```python
detector.loadModel()
list = detector.detectObjectsFromImage(
    input_image = 'Image 01.jpg', # Задание имени входного файла
    output_image_path = 'Image 01_out_RetinaNet.jpg', # Задание имени выходного файла
    minimum_percentage_probability = 50 # Задание порога отсечения
)
```

```python
Image(filename='Image 01_out_RetinaNet.jpg')
```

```python
detector = ObjectDetection()
detector.setModelTypeAsYOLOv3() # Модель YOLO  
detector.setModelPath("yolo.h5")  
detector.loadModel()
```

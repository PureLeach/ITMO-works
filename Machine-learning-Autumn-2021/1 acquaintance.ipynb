1)	Тема бакалаврской ВКР: «Использование технологий промышленного интернета вещей (IIoT)	в системах контроля и диагностики автоматизированного электропривода»

В рамках выпускного квалификационного проекта спроектировано электронное устройство для снятия параметров автоматизированного электропривода и их передачи по беспроводным сетям с помощью технологий промышленного интернета вещей. Для разработки устройства была проведён выбор элементной базы, созданы функциональная и принципиальные схемы. Итогом разработки стало проектирование печатной платы и корпуса электронного устройства.
В свою очередь для микроконтроллера ESP8266, установленного в устройство, разработана программа на языке программирования Lua. 
Для создания сервиса контроля и диагностики автоматизированного электропривода использовалась среда разработки проектов интернет вещей Node-RED. С помощью данного инструмента был создан сервис, который позволяет выполнять следующие функции: учёт базы данных электроприводов, установленных на предприятии, сбор и анализ полученных данных по протолку MQTT, оценка состояния агрегатов, определение имеющихся дефектов, формирование отчётов и вывод результатов диагностики.
Для обеспечения информационной безопасности была предусмотрена система защиты данных на базе криптографических протоколов TLS и SSL. 
Кроме разработанного сервиса, устройство можно подключить к SCADA-системам, которые поддерживают протокол MQTT, такие как WinCC Open Architecture, MasterSCADA 4D и другие. С каждым годом количество систем, поддерживающих протокол MQTT растёт.
Устройство может быть применено как в промышленности, так и в учебных целях для изучения новых технологий в области промышленного интернета вещей.

Для проверки качества спроектированного устройства и разработанных программ была использована методика диагностики асинхронного короткозамкнутого электропривода по результатам измерения параметров рабочего режима. В программу был добавлен алгоритм данного метода.
Внедрение разработанного устройства и программного сервиса обеспечит беспрерывный автоматический контроль за состоянием электрооборудования. 
В результате этого будет сокращено количество непредвиденных случаев вывода из строя электрооборудования, приводящие к нарушению непрерывности технологических процессов с последующим браком продукции, затратам на восстановление и ремонт электродвигателей, а также на восстановление нормальных технологических процессов производства. В свою очередь внедрение системы позволит снизить до минимума ущерб от этих последствий за счет раннего обнаружения зарождающихся дефектов и повышению безопасности на производстве.


2)	Тема магистерской ВКР: «Разработка системы мониторинга состояния высоконагруженной вычислительной стойки»

1	Результаты выполнения проекта
В результате выполнения проекта должен быть создан и интегрирован на ООО «СТЦ» прототип программно-аппаратного модуля для мониторинга состояния высоконагруженной стойки», включающий в себя:
1.1	Комплекс программных средств в составе:
1.1.1	Программное обеспечение модуля мониторинга реализующие бизнес-логику отдельного устройства мониторинга, человеко-машинный интерфейс, сбор данных и визуализацию информации с вычислительных стоек. 
1.1.2	Комплекс программных средств центра мониторинга и контроля, реализующих бизнес-логику дата-центра: обработку данных с вычислительных стоек, агрегирование, визуализацию данных для анализа, сбор и передачу данных управления на локальные площадки, прогнозирование работоспособности оборудования с использованием когнитивных технологий и машинного обучения.
1.2	Аппаратный комплекс для обеспечения функционирования прототипа.

2	Назначение продукции
2.1	Модуль мониторинга состояния высоконагруженной вычислительной стойки предназначен для отслеживания технологических параметров вычислительной стойки, находящейся на территории ЦОД, обеспечения безопасности эксплуатации оборудования, оптимизации операционной деятельности, автоматизации отчётности, накопления информации о произошедших отклонениях для ретроспективного анализа и возможности построения регрессионных моделей, увеличения оперативности отработки отклонений, снижения затрат и повышения эффективности.
2.2	Программное обеспечение модуля мониторинга предназначено для реализации бизнес-логики дата-центра, сбора и визуализации технологических параметров, полученных от клиентов и от внешних датчиков. 
2.2.1	К технологическим параметрам клиента относятся данные, которые можно получить напрямую от оборудования с помощью специализированного ПО:
–	температура CPU (°C), 
–	нагрузка CPU (%),
–	скорость вращения вентиляторов (об/мин)
–	доступность портов
–	наличие свободной оперативной памяти на сервере
–	мониторинг здоровья жесткого диска по SMART и состояния RAID массива
–	мониторинг появления новых ошибок в логе сервера
2.2.2	К параметрам получаемых от внешних датчиков, закреплённых внутри вычислительной стойки относятся:
–	температура внутри стойки (°C), 
–	влажность внутри стойки (%)
2.3	Комплекс программных средств центра мониторинга и контроля предназначен для реализации бизнес-логики дата-центра для обеспечения работоспособности и эффективности работы серверов.
2.4	Аппаратный комплекс предназначен для реализации прототипа модуля мониторинга на предприятии ООО «СТЦ».

3	Технические требования к программно-аппаратному комплексу

3.1	Состав продукции
В состав разрабатываемого прототипа должны входить:
•	Программные средства, перечисленные в п. 1.1.
•	Аппаратные средства, указанные в п. 1.2.

3.2	Требования к функциональным характеристикам
Требования к функциональным характеристикам, приведённые в настоящем разделе Технического Задания являются общими. Дополнительные функциональные требования к каждому компоненту модуля мониторинга определяются соответствующими техническими требованиями (ТТ) по результатам выполнения 1-го этапа проекта.
3.2.1	Программное обеспечение модуля мониторинга должно быть программно независимым от используемых средств центра мониторинга и контроля, и реализовывать свои функции на основе данных полученных от внешних и внутренних датчиков вычислительной стойки. Необходимо реализовать набор человеко-системных интерфейсов, в том числе:
3.2.1.1	Рабочая область должна представлять собой интерфейс визуализации состояния оборудования в режиме реального времени .
3.2.1.2	На экране оператора должен быть вывод информации в виде графиков с возможностью переключения промежутка времени.
3.2.1.3	Необходимо реализовать переключение вывода информации о каждом сервере установленным в стойку. 
3.2.2	Период обновления данных не должен быть чаще 1 секунды.
3.2.3	Количество внешних датчиков должно быть не менее 6 для каждого параметра (температура и влажность) на одну стойку.
3.2.4	Датчики закреплённые снизу должны находится на расстоянии 4U от основания стойки.
3.2.5	Датчики закреплённые сверху должны находится на расстоянии 44U от основания стойки.
3.2.6	Экран и элементы управления должны быть встроены в металлический корпус устройства.
